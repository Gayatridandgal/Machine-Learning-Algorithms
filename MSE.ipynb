```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Example data (replace with your actual data)
X = np.array([1, 2, 3, 4, 5])  # Input features
Y = np.array([2, 4, 6, 8, 10])  # Target values

# Initialize parameters
m = 0
c = 0

L = 0.0001  # Learning rate
epochs = 2000  # Number of iterations

n = float(len(X))  # Number of elements in X

# List to store MSE for each epoch
mse_list = []

# Perform gradient descent
for i in range(epochs):
    Y_pred = m * X + c  # Predicted Y values
    
    # Compute gradients
    D_m = (-1/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-1/n) * sum(Y - Y_pred)  # Derivative wrt c
    
    # Update parameters
    m = m - L * D_m
    c = c - L * D_c
    
    # Calculate MSE and store it
    mse = mean_squared_error(Y, Y_pred)
    mse_list.append(mse)

# Plotting the MSE vs Epochs
plt.plot(range(epochs), mse_list)
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Learning Curve")
plt.show()
```

```python
plt.plot(range(epochs), mse_list, color='b', linewidth=2)  # Smooth curve with thicker line
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Learning Curve (MSE vs Epochs)")
plt.grid(True)  # Add grid for better readability
plt.show()
```
